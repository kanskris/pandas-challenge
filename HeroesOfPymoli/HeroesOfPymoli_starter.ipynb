## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_count=purchase_data["Gender"].value_counts()
gender_count
```

```python
gender_pct=purchase_data["Gender"].value_counts(normalize=True)*100
gender_pct
```

```python
gender_dem= pd.concat([gender_count,gender_pct],axis=1,keys=('Total Count','Percentage of Players'))
gender_dem['Percentage of Players']=pd.Series(["{0:.2f}%".format(val) for val in gender_dem['Percentage of Players']],index = gender_dem.index)
gender_dem
```

```python
pa_sum_price=purchase_data.drop(['SN','Age','Item ID','Item Name','Purchase ID'], axis=1)
pa_sum_price
```

```python
pa_sum_price= pa_sum_price.groupby('Gender')
pa_sum_it_up = pa_sum_price.sum()
pa_sum_it_up.head()
```

```python
pa_final=pd.concat([pa_count_it_up,pa_avg_it_up,pa_sum_it_up],axis=1, keys=('Count of','Avg of Purchase','Total Purchase'))
pa_final
```
